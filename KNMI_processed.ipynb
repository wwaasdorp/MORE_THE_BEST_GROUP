```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats 
import pandas as pd
```

```python
file_path = 'data/KNMI_hvh.txt'

df = pd.read_csv(
    file_path,
    skiprows=51,  # Skip non-data rows
    header=0,  # The header row starts with '# STN,'
    sep=',',  # Columns are comma-separated
    skipinitialspace=True  # Skip spaces after commas
)

# Ensure the date column is identified (e.g., 'YYYYMMDD')
# Replace 'DATE_COLUMN' with the actual name of your date column
df['YYYYMMDD'] = pd.to_datetime(df['YYYYMMDD'], format='%Y%m%d', errors='coerce')

# Set the date column as the index
df.set_index('YYYYMMDD', inplace=True)
```

```python
df_pressure = df[['PG', 'PX', 'PXH', 'PN', 'PNH']].copy()
```

```python
plt.plot(df_pressure.index, df_pressure['PG'], label='PG')

plt.xlabel('Date')
plt.ylabel('Pressure (hPa)')
plt.title('Pressure in Hoek van Holland')
```

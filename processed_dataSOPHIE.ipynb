```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('Meetreeks_HVH.txt', sep="\s+", header=None, names=["Date", "Time", "Water_Level"])
# Access the columns
dates = df["Date"]
times = df["Time"]
water_levels = df["Water_Level"]
df.head()
```

```python
df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])
```

```python
plt.plot(df['Datetime'], df['Water_Level'], label="Water Level")
plt.ylabel('Water level (cm)')
plt.title('Observed waterlevels in 1888-2020 at Hoek van Holland');
```

```python
df['Y'] = df['Datetime'].dt.year
yearly_max = df.groupby('Y')['Water_Level'].max()
yearly_max.plot(title='Yearly maximum water levels at Hoek van Holland', xlabel='Time', ylabel='Water Level (cm)');
```

In the graph, the yearly maxima of waterlevels at Hoek van Holland are plotted. It can be observed that the maximum areusually around 200-300 cm. At one moment, a very high peak was observed. This is during the flood disaster of 1953, which flooded large parts of Zeeland and Zuid-Holland. 

```python
POT200 = pd.DataFrame()
POT200 = df.loc[(df['Water_Level'] >= 200)]
plt.plot(POT200['Datetime'], POT200['Water_Level'], label="Water Level")
plt.xlabel('Time')
plt.ylabel('Water level (cm)')
plt.title('Extreme water levels at Hoek van Holland, POT threshold = 200 cm');
```

In the upper graph, Peak Over Threshold is used to obtain extreme values for the water levels at Hoek van Holland. It looks like the amount of observations increases around 1970. This means that extreme water levels > 200 cm occur more often after 1970. To check if this is true, the yearly amount of exceedences is plotted in the next graph.
